```python
pokemon_dict = {}
for i in range(len(pokemon_data['name'])):
    key = pokemon_data['name'][i].upper()
    val = [
        int(pokemon_data['pokedex_number'][i]),
        pokemon_data['type_1'][i],
        ["None" if type(pokemon_data['type_2'][i])==float else pokemon_data['type_2'][i]][0],
        float(pokemon_data['height_m'][i]),
        float(pokemon_data['weight_kg'][i]),
        int(pokemon_data['generation'][i]),
        int(pokemon_data['total_points'][i])
    ]
    pokemon_dict[key] = val
pokemon_dict
```
